## Types of neural network architectures


### Feed-forwad neural networks

- These are the commonest type of neural network in pratical applications
    - The first layer is the input and last layer is the output
    - If there is more than one hidden layer, we call them "deep" neural network
- They compute a series of transformations that change the similarities between cases
    - The activities of the neurons in each layer are a non-linear function of the activities in the layer below

### Recurrent neural network

- These have directed cycles in their connection graph
    - That means you can sometimes get back to where you started by following the arrows
- They can have complicated dynamics and this make them very difficult to train
    - There is a lot of interest at present in finding efficient ways of training recurrent nets
- They are more biological realistic

(Recurrent nets with multiple layers are just a special case that has some of hidden -> hidden connections missing)

### Recurrent neural networks for modeling sequences

- Recurrent neural networks are a very natural way to model sequential data
    - They are equivalent to very deep nets with one hidden layer per time slice
    - Except that they use the same weights at every time slice and they get input at every time slice
- They have the ability to remember informationin their hidden state for a long time
    - But it's very hard to train them to use this potential

### Symmetrically connected networks

- These are like recurrent networks, the the connection between units are simmetrical (they have the same weight int both directions).
    - John Hopfield (and others) realized that symmetric networks are much easier to analyze than recurrent networks
    - They are also more restricted in what they can do, because the obey an energy function
        - For example, they cannot model cycles
        
- Symmetrically connected nets without hidden units are called "Hopfield nets"

## Perceptrons: the first generation of neural networks

### The standard paradigm for statistical pattern recognition

1. Convert raw input vector into a vector of features activations
- use hand-written programs based on commons sense to definethe features

2. Define how to weight each of feature activations to get a single scalar quantity
3. If this quantity is above some threshold, decide that the input vector is a positive example of the target class

### The history of perceptrons

- They were popularised by Frank Rosenblatt in the early 1960's
    - They appeared to have a very powerful learning algorithm
    - A lot of grand claims were made for what they could learn to do
- In 1969, Minsky and Papert published a book called "Perceptrons" that analysed what they could do and showed their limitations
    - Many people thought these limitations applied to all neural network models
- The perceptron learning procedure is widely used today for taskswith enormous feature vectors that contain many milions of features

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def binary_threshold(z):
    return 1 if z >= 0 else 0

def binary_neuron(b, x, w = [-1,1,1], threshold = binary_threshold):
    if len(x) != len(w):
        raise ValueError("length of x and w are diferent")
    z = b + sum([x[i] * w[i] for i in range(len(x))])
    y = threshold(z)
    return y, z
```

```python
assert binary_neuron(1, [1,0,1]) == (1,1)
```

```python
assert binary_neuron(1, [1,-1,1]) == (1,0)
```

```python
assert binary_neuron(1, [1,-2,1]) == (0,-1)
```

```python
plt.plot([-1,0,0,1],[0,0,1,1])
plt.axis([-1,1,-0.1,1.1])
plt.grid(True)
plt.show()
```

### How to learn biases using the same rule as we use for learning weights

- a threashold is equivalent to having a negative bias
- we can avoid having to figure out a separated learning rule for the bias by using a trick:
    - A bias is exactly equivalent to a weight on an extra input line that always has an activity of 1
    - We can now learn a bias as if it were a weight 

### The perceptron convergence procedure: Training binary output neurons as classifiers

- Add an extra component with value 1 to each input vector. The "bias" weight on this component is minus the threshold. Now we can forget the threshold.
- Pick training cases using any policy that ensures that every training case will keep getting picked
    - If the output unit is correct, leave its weights alone
    - If the output unit incorrectly outputs a zero, add the input vector to the weight vector
    - If the output unit incorrectly outputs a 1, subtract the input vector from the weight vector
- This is guaranteed to find a set of weights that gets the right answer for all the training cases **if any such set exists**

```python
def learn_biases(b, resulted_y, expected_y, actual_w, x):
    if resulted_y == expected_y:
        return expected_y
    
    w_b = list(actual_w)
    w_b.append(b)
    
    x_1 = list(x)
    x_1.append(1)
    
    # (w,b) <- (w,b) - (x,1)
    return (np.subtract(w_b, x_1).tolist() if resulted_y > expected_y else np.add(w_b, x_1).tolist())

assert learn_biases(b=1, resulted_y=0, expected_y=1, actual_w=[1,1,1], x=[1,1,1]) == [2,2,2,2]
```

```python
x=[0.5,-0.5]
expected_y=0
actual_w=[2,-1]
b=0.5
resulted_y = binary_neuron(b=b, x=x, w=actual_w)

assert resulted_y == (1, 2.0)

assert learn_biases(b, resulted_y[1], expected_y, actual_w, x) == [1.5, -0.5, -0.5]
```

### Weight-space

- This space has one dimension per weight
- A point in the space represents a particular setting of all the weights.
- Assuming that we have eliminated the threshold, each training case can be represented as a hyperplane through the origin.
    - **Note:** points in the space correspond to weight vectors and training cases correspond to planes
    - To particular training case, the weights must lie on one side of this hyper-plane to get the answer correct.


**Note:** In the weight space view, the weights represent points while the inputs represent planes. Another term for what the inputs represent is **Constraints**

We can think of the inputs as partitioning the space into two halves. Weights lying in one half will get the answer correct while on the other half they will give the incorrect answer (which half is determined by the output class). That is, the inputs will *constrain* the set of weights that give correct classification results.

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
from pylab import *

alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)

phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T
```

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1, 1, 1, projection='3d')

ax.text(0.04,0.04,0.04, "right answer")

ax.text(-0.04,-0.04,-0.04, "wrong answer")

ax.text(2,2,0, "inputs or constraints", color='white')

p = ax.plot_surface(X, Y, 0, rstride=4, cstride=4)

    
show()
```

## Why the learning works

- consider the squared distance distance $d^2_a + d^2_b$ between any feasible weight vector and the current weight vector
    - **Hopeful claim:** every time the perceptron makes a mistake, the learning algorithm moves the current weight vector closer to all feasible weight vectors

```python
plt.plot([-1,0,1],[0,0,0])
plt.axis([-1,1,-1,1])
plt.grid(True)

plt.text(x=0.7, y=0.05, s='right')
plt.text(x=0.7, y=-0.1, s='wrong')

plt.scatter(x=0.5, y=0.5, c='green')
plt.text(x=0.5, y=0.5, s='feasible')

plt.scatter(x=0, y=-0.3, c='red')
plt.text(x=0, y=-0.3, s='current')

plt.scatter(x=-0.5, y=0.2, c='yellow')
plt.text(x=-0.5, y=0.2, s='Problem case: the weight \n vector may not get closer \n to this feasible vector')

plt.show()
```

- So consider "generously feasible" weight vectors that lie within the feasible region by margin at least as great as the lenght of the input vector that defines each constraint place
    - every time the perceptron makes a mistake, the squared distance to all of these generously feasible weight vectors is always decreased by at least the squared lenght of the update vector.

```python
plt.plot([-1,0,1],[0,0,0])
plt.plot([-1,0,1],[0.4,0.4,0.4], '--')
plt.axis([-1,1,-1,1])
plt.grid(True)

plt.text(x=0.0, y=0.3, s='margin', rotation='vertical')

plt.text(x=0.7, y=0.05, s='right')

plt.scatter(x=0.5, y=0.5, c='green')

plt.scatter(x=0, y=-0.3, c='red')

plt.scatter(x=-0.5, y=0.2, c='yellow')

plt.show()
```

### Informal sketch of proof of convergence
- Each time the perceptron makes a mistake, the current weight vectors moves to decrease its squared distance from every weight vector in the "generously feasible" region
- The squared distance decreases by at least the squared lenght of the input vector
- So after a finite number of mistakes, the weight vector must lie in the feasible region *if this region exists.*

## What perceptrons can't do

- If you are allowed to choose the features by hand and if you use enough features, you can do almost anything.
    - For binary input vectors, we can have a separate feature unit for each of the exponentially many binary vectors and do we can make any possible discrimination on binary input vectors.
        - This type of table look-up won't generalize
- But once the hand-coded features have been determined, there are very strong limitations on what a perceptron can learn

- A binary threshold output unit cannot event tell if two single bit features are the same!
    - Positive cases (same): (1,1) -> 1; (0,0) -> 1
    - Negative cases (different): (1,0) -> 0; (0,1) -> 0
- The four input-output pairs give four inequalities that are impossible to satisfy:

$w_1 + w_2 \geq \theta, 0 \geq \theta$

$w_1 \lt \theta, w_2 \lt \theta$

### A geometric view of what binary threshold neurons cannot do

Imagine "data-space" in wich the axes correspond to components of an input vector.
- Each input vector is a point in this space
- A weight vector defines a plane in data-space
- The weight plane is perpendicular to the weight vector and misses the origin by a distance equal to the threshold


```python
plt.plot([-0.5,1.5],[0.5,0])
plt.axis([-0.5,1.5,-0.5,1.5])
plt.grid(True)

plt.text(x=0.7, y=0.3, s='output=1')
plt.text(x=0.7, y=0.05, s='output=0')

plt.text(x=0.2, y=0.2, s='weight plane', rotation='-10')

plt.scatter(x=0, y=0, c='green')

plt.scatter(x=1, y=1, c='green')

plt.scatter(x=0, y=1, c='red')

plt.scatter(x=1, y=0, c='red')

plt.xlabel("The positive and negative cases cannot be separated by a plane")

plt.show()
```

### Discriminating simple patterns under translation with wrap-around

- Suppose we just use pixels as the features
- Can a binary threshold unit discriminate between different patterns that have the same number of on pixels?
    - No if the patterns can translate with wrap-around!

### Sketch of a proof that a binary decision unit cannot discriminate patterns with the same number of on pixels (assuming translation with wraparound)

- For pattern A, use training cases in all possible translations.
    - Each pixel will be activated by 4 different translations of pattern A
    - So the total input received by the decision unit over all these patterns will be four times the sum of all the weights
- For pattern B, use training cases in all possible translations.
    - Each pixel will be activated by 4 different translations of pattern B
    - So the total input received by the decision unit over all these patterns will be four times the sum of all the weights
- But to discriminate correctly, every single case of pattern A must provide more input to the decision unit than every single case of pattern B

**Note:** Suppose we have three patterns that belong to two classes. The first class contains patterns with 4 pixels on while the second class contains patterns with either 1 or 3 pixels on. Can a binary decision unit classify these correctly if we allow for translations with wraparound?

If we set the weight coming from each pixel to be 1, and the bias to be -3.5, then any example with 3 pixels on will receive a total activations of -0.5, any example with 1 pixel on will receive a total activation of -2.5 and any example with 4 pixels on will receive an activation of 0.5. This correctly classifies the dataset.

### Why this result is devastating for Perceptrons

- The whole point of pattern recognition is to recognize patterns despite transformation like translation
- Minsky an Papert's "Group Invariance Theorem" says that the part of a Perceptron that learns cannot learn todo this if the transformation form a group
    - Translations with wrap-around form a group
- To deal with such transformations, a Perceptron needs to use multiple features units to recognize transformations of informative sub-patterns.
    - So the tricky part of pattern recognition must be solved by the hand-coded features dectectors, not the learning procedure

### Learning with hidden units

- Networks without hidden units are very limited in the input-output mappings they can learn to model
    - More layers of linear units do no help. Its still linear
    - Fixed output non-linearities are not enough
- We need multiple layers of *adaptative,* non-linear hidden units. But how can we train such nets?
    - We need an efficient way of adapting *all* the weights, not just the last layer. This is hard.
    - Learning the weights going into hidden units is equivalent to learning features
    - This is difficul because nobody is telling us directly what the hidden units should do
